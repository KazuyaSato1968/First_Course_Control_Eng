```python
#from control.matlab import *
from control import matlab
from matplotlib import pyplot as plt
from scipy import arange 
```

```python
num1 = [0, 1] 
num2 = [0, 20] 
num3 = [0, 50] 
den1 = [1,1]
den2 = [20,1]
den3 = [50,1]
sys1 = matlab.tf(num1, den1) 
sys2 = matlab.tf(num2, den2) 
sys3 = matlab.tf(num3, den3) 
```

```python
(y1, T1) = matlab.initial(sys1, X0 = [1.0], T = arange(0, 50, 0.01))
(y2, T2) = matlab.initial(sys2, X0 = [1.0], T = arange(0, 50, 0.01))
(y3, T3) = matlab.initial(sys3, X0 = [1.0], T = arange(0, 50, 0.01))

plt.plot(T1, y1, label = "T = 1")
plt.plot(T2, y2, label = "T = 20")
plt.plot(T3, y3, label = "T = 50")
plt.xlim([0,50])
plt.grid(color='gray')
plt.xlabel("time   t[s]")
plt.ylabel("output")
plt.legend()
plt.show()
```
